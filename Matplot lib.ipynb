```python
import matplotlib.pyplot as plt
# creating linear plot(it is like a graph we have two axis x and y and we join two cordinates using line )
x=[1,2,5,8,9]
y=[20,22,25,29,30]
plt.plot(x,y)
# in above we are seeing some lines are written above plot if we want to remove these lines only want to see the plot
plt.show()
```

```python
x=[1,2,5,8,9]
y=[20,22,25,29,30]
# if we want to give colour to the line of graph or plot
c='r'#we write r so red colour will be given
plt.plot(x,y,c)
# in above we are seeing some lines are written above plot if we want to remove these lines only want to see the plot
plt.show()
```

```python
x1=[5,2,9,4,7]
y1=[10,5,8,4,2]
plt.plot(x1,y1)
```

```python
# now we are adding marker at the place where cordinates joins
m=[5,9,3,1]
n=[9,3,4,10]
plt.title("Linear")
plt.plot(m,n,marker="o")
# we can see axis differntly with marker
```

```python
m=[5,9,3,1]
n=[9,3,4,10]
plt.plot(m,n,marker="o")
# we can see axis differntly with marker
plt.plot(m,marker="<")
plt.plot(n,marker="<")
```

```python
# bar plot
b1=[1,2,3,4]
b2=[5,6,7,8]
plt.bar(b1,b2)
```

```python
# now i will add coours in bars
b1=[1,2,3,4]
b2=[5,6,7,8]
c=['r','g','b','y']
plt.bar(b1,b2, color = c)
plt.show()
```

```python
# now creatin bar graph horizentaly
bh=['a','b','c']
bh2=[1,10,3]
plt.barh(bh,bh2)
```

```python
# now creating bar graph from arrays
import numpy as np
ar1=np.array(['A','B','C','D'])
ar2=np.array([10,20,30,40])
c=['r','g','b','y']
# we can write the code of color="#asa234" at the place of c
plt.barh(ar1,ar2,color=c)
```

```python
# we can set the width of bars of bar graph
ar1=np.array(['A','B','C','D'])
ar2=np.array([10,20,30,40])
# we can write the code of color="#asa234" at the place of c
plt.bar(ar1,ar2,color="#4caf50",width=0.1)
```

```python
# if we are representing bar grraph horizently we can set height
ar1=np.array(['A','B','C','D'])
ar2=np.array([10,20,30,40])
plt.title("Bar")
c=['r','g','b','y']
plt.barh(ar1,ar2,color=c,height=0.1)
```

```python
# Scatter Plot
b1=[1,2,3,4]
b2=[5,6,7,8]
plt.title("Scatter")
plt.scatter(b1,b2)
plt.show()
```

```python
b1=[1,2,3,4]
b2=[5,6,7,8]
plt.title("Scatter")
# now giving label on axis of plot
plt.xlabel('Month')
plt.ylabel('Number')
plt.scatter(b1,b2,color='r')#now we added color
plt.show()
```

```python
b1=[1,2,3,4]
b2=[5,6,7,8]
plt.title("Scatter")
# now giving label on axis of plot
plt.xlabel('Month')
plt.ylabel('Number')
c=['r','g','b','y']
plt.scatter(b1,b2,color=c)#now we added dfferent color for cordinates
plt.show()
```

```python
b1=[1,2,3,4]
b2=[5,6,7,8]
plt.title("Scatter")
# now giving label on axis of plot
plt.xlabel('Month')
plt.ylabel('Number')
c=['r','g','b','y']
plt.scatter(b1,b2,color=c,s=150)#now we added dfferent color for cordinates,no ichange the size of scatter
plt.show()
```
